```python
df.fillna(method='bfill')
```

```python
df.fillna(method='ffill',axis='columns')
```

```python
df.fillna(method='ffill',limit=2)
```

```python
df.plot.bar(x='EST', y='WindSpeedMPH')
```

```python
df.interpolate()
```

```python
df1 = df[['EST','WindSpeedMPH']]
df2 = df[['WindSpeedMPH']].rename(columns={"WindSpeedMPH": "WindSpeedMPH_intr"}).interpolate()

df3 = pd.concat([df1, df2], axis=1).set_index('EST')
ax = df3.plot.bar()
```

**merge**

```python
import pandas as pd
df1 = pd.DataFrame({
    "city":["new york", "chicago", "orlando"],
    "temperature" : [21,14,35]
})
df1
```

```python
df2 = pd.DataFrame({
    "city":[ "chicago","new york", "orlando"],
    "humidity" : [65,68,75]
})

df2
```

```python
df3 = pd.merge(df1,df2,on='city')
df3
```

```python
df1 = pd.DataFrame({
    "city":["new york", "chicago", "orlando","karimnagar"],
    "temperature" : [21,14,35,101]
})

df2 = pd.DataFrame({
    "city":[ "chicago","new york", "orlando","hyderabad"],
    "humidity" : [65,68,75,99]
})

df3 = pd.merge(df1,df2,on='city')
df3
```

```python

df3 = pd.merge(df1,df2,on='city', how="outer")
df3
```

```python
df3 = pd.merge(df1,df2,on='city', how="left")
df3
```

```python
df3 = pd.merge(df1,df2,on='city', how="right")
df3
```

```python
df3 = pd.merge(df1,df2,on='city', how="outer",indicator=True)
df3
```
