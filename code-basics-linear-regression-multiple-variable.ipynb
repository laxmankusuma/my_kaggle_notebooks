**mean:** 

(sum of all) / (count of all)

**median**

The median is the same as the 50 percentile.

i.e arrange all in a row take middle one is the median. i.e right side 50% and left side 50%

 **Q) when we will go for mean and when we will go for median**


for example1 : mean of 1 2 3 is (1+2+3)/3=2

for example2 : mean of 1 2 3 4 4 5 6 6 7 7 1000 is (1+2+3+4+4+5+6+6+7+7+1000)/11=95

median of 1 2 3 4 4 5 6 6 7 7 1000 is 5

so in first case mean is better to fill missing values

in second case median is better to fill missing values

**25%**

arrange all values in order then it has 25% of the values left side and 75% of the values right side

its also known as first quadrant (Q1)
