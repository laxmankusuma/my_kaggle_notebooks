**Categorical variables**

1. **Nominal**(order not follows) (ex: male,female)
2. **Ordinal**(order follows) (ex: degree, masters, phd)

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/patilankita79/ML_DataPreprocessing/master/ML_DataPreprocessing/Data.csv")
```

```python
df
```

```python
dummies = pd.get_dummies(df['Country'])
dummies
```

```python
merged = pd.concat([df,dummies], axis='columns')
merged
```

```python
#no need one columns in dummies
final_df = merged.drop(['Country','France'],axis='columns')
final_df
```
