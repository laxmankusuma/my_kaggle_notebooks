```python
import numpy as np
```

```python
company = ['google','amazon','netflix','facebook']
revenue = [90,78,68,85]
```

```python
ypos = np.arange(len(company))
ypos
```

```python
plt.bar(ypos,revenue)
```

```python
plt.title('revenue analysis')
plt.ylabel('revenue in bns')
plt.xticks(ypos,company)
plt.bar(ypos,revenue)
```

```python
profit = [40,2,32,8]
plt.title('revenue analysis')
plt.ylabel('revenue in bns')
plt.xticks(ypos,company)
plt.bar(ypos,revenue,label='revenue')
plt.bar(ypos,profit,label='profit')
plt.legend()
```

```python
profit = [40,2,32,8]
plt.title('revenue analysis')
plt.ylabel('revenue in bns')
plt.xticks(ypos,company)
plt.bar(ypos-0.2,revenue,label='revenue')
plt.bar(ypos+0.2,profit,label='profit')
plt.legend()
```

```python
profit = [40,2,32,8]
plt.title('revenue analysis')
plt.ylabel('revenue in bns')
plt.xticks(ypos,company)
plt.bar(ypos-0.2,revenue,width=0.4,label='revenue')
plt.bar(ypos+0.2,profit,width=0.4,label='profit')
plt.legend()
```

```python
profit = [40,2,32,8]
plt.title('revenue analysis')
plt.yticks(ypos,company)
plt.barh(ypos-0.2,revenue,label='revenue')
plt.barh(ypos+0.2,profit,label='profit')
plt.legend()
```

**pie chart**

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["home","food","car","other","electronic"]
```

```python
plt.pie(exp_vals,labels=exp_labels)
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels)
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1,autopct='%0.1f%%')
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1,autopct='%0.1f%%',shadow=True)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1,autopct='%0.1f%%',shadow=True,explode=[0,0,0,0,0])
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1,autopct='%0.1f%%',shadow=True,explode=[0,0.1,0.1,0,0])
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1,autopct='%0.1f%%',shadow=True,explode=[0,0.1,0.1,0,0],startangle=180)
# plt.show()
plt.savefig('piechart.png',bbox_inches='tight')
```
